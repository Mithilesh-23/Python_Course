```python
import pandas
import matplotlib.pyplot as plt
```

```python
url="https://raw.githubusercontent.com/Mithilesh-23/Python_Course/refs/heads/main/Dataset/Flipkart_Mobiles_Sale.csv?token=GHSAT0AAAAAADFH4O6WPJRRRSPUPBBI24TQ2DJDGAA"
df=pandas.read_csv(url)
print(df)
```

```python
topBrand=df['Brand'].value_counts().head(10)
print(topBrand)
```

```python
labels=topBrand.index
plt.pie(topBrand, labels=labels, autopct='%1.1f%%')
plt.title("Brand and their sales")
plt.show()
```

```python
topColor=df['Color'].value_counts().head(10)
print(topColor)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(topColor.index, topColor.values, color='lightblue', edgecolor='blue')
plt.title("Color wise Sale")
plt.xlabel("Color Name,", fontsize=10)
plt.ylabel("Sales Count")
plt.title("Color Wise Sales", fontsize=15, fontweight='bold')
plt.grid(axis='y')
plt.xticks(rotation=90)
for i, value in enumerate(topColor.values):
  plt.text(i, value + 2, str(value), ha='center',va='bottom', fontsize=10)
plt.show()
```

```python
topModel=df['Model'].value_counts().head(20)
print(topModel)
```

```python
plt.figure(figsize=(15, 6))
plt.bar(topModel.index, topModel, color='grey', edgecolor='black')
plt.title("Mobile Mobile Sale", color='Red', fontsize='15')
plt.xlabel("Model Name", color='blue')
plt.ylabel("Modle sale", color='blue')
plt.xticks(rotation=90)
plt.grid(axis='y')
for i, value in enumerate(topModel.values):
  plt.text(i, value + 0.3, str(value), ha='center', va='bottom')
plt.show()
```

```python
import numpy as np
```

```python
saleComparision=df.groupby('Brand')[['Original Price', 'Selling Price']].sum()
print(saleComparision)
```

```python
brand=saleComparision.index
print(brand)
x=np.arange(len(brand))
width=0.4
print(x)

plt.figure(figsize=(15, 7))
plt.bar(x-width/2, saleComparision['Original Price'], width, label='Original Price')
plt.bar(x+width/2, saleComparision['Selling Price'], width, label='Selling Price')
plt.xlabel("Brand Name")
plt.ylabel("Price in Rs")
plt.title("Brand Sale Comparision")
plt.xticks(x, brand, rotation=90)
for i, value in enumerate(saleComparision['Original Price']):
  plt.text(i, value+1, str(value), ha='right', va='bottom', fontsize=7, color='blue')
for i, value in enumerate(saleComparision['Selling Price']):
  plt.text(i, value+3, str(value), ha='left', va='baseline', fontsize=7, color='orange')
plt.grid(axis='y')
plt.legend()
plt.show()
```
